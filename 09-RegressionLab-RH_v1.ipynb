this set of scatterplots gives us an idea of the pairwise relationships present in the dataset

```python
sns.pairplot(x, size = 1.2, aspect=1.75)
```

```python
x.corr()
```

this linear fit represents the "full model"; eg, the fit with all of the independent variables included

```python
lm = smf.ols('Y ~ X1 + X2 + X3 + X4 + X5 + X6', data=x)
fit = lm.fit()
print fit.summary()
```

remove feature w/ lowest (abs) t score

```python
#remove X5
fit2 = smf.ols('Y ~ X1 + X2 + X3 + X4 + X6', data=x).fit()
print fit2.summary()
```

note R-sq decreases slightly, but adj R-sq increases slightly

--> increasing bias, decreasing variance

ditto

```python
#x4 is not significant, remove it
fit3 = smf.ols('Y ~ X1 + X2 + X3 + X6', data=x).fit()
print fit3.summary()
```

```python
#x2 is not significant, remove it
fit4 = smf.ols('Y ~ X1 + X3 + X6', data=x).fit()
print fit4.summary()
```

stopping criteria met: all featuers have `|t| > 1`

$\rightarrow$ optimal bias-variance point reached

$\rightarrow$ Residual standard error (RSE) minimized

```python
'''
note other potential errors exist -- 

DW is okay
Durbin-Watson = autocorrelation (time-lag): 
 0 = min ==> positive autocorrelation 
 2 = mid ==> no autocorrelation in sample
 4 = max ==> negative autocorrelation 
 http://www.investopedia.com/terms/d/durbin-watson-statistic.asp

JB/Skew/Kurtosis is ... 
Skewness-Kurtosis (Jarque-Bera) Stat
Skew - goal is 0
Kurtosis - goal is 3
JB = tests if error residuals are normally distributed

Multicolinearity
http://en.wikipedia.org/wiki/Multicollinearity
 data.corr()
     ==> X1 ~ x3 = 0.60, x3 ~ x6 = .53

Heteroskedacity
http://en.wikipedia.org/wiki/Heteroscedasticity
'''
x.corr()
```

```python
#given multicolinearity between x3 and x1 and x6, dropping x3
#x3 is colinear to x1 & x6
fit5 = smf.ols('Y ~ X1 + X6', data=x).fit()
print fit5.summary()
```

```python
#sns.distplot(fit4.resid, rug= 'TRUE')
sns.distplot(fit4.resid, rug= 'TRUE', bins=10)
```

```python
#sns.distplot(fit5.resid, rug= 'TRUE')
sns.distplot(fit5.resid, rug= 'TRUE', bins= 10)

#sns.distplot?
```

```python
plt.figure(figsize=(12,8))
_ = stats.probplot( fit4.resid, dist="norm", plot=plt)
```

```python
plt.figure(figsize=(12,8))
_ = stats.probplot( fit5.resid, dist="norm", plot=plt)
```

note this model is weaker (lower $Adj. R^2$)

```python
#but... it is stronger than than teh 'Y ~ X1 and X6' 
fit6 = smf.ols('Y ~ X1', data=x).fit()
print fit6.summary()
```
