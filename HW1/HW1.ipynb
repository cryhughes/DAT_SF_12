####7. Show 10 of the sidewalk cleanings in the Haight Ashbury neighborhood

```python
#YOUR CODE HERE
pd.read_sql( 'SELECT * FROM sf_311 '
            'WHERE category LIKE "%Sidewalk Cleaning%" AND neighborhood = "Haight Ashbury" '
            'LIMIT 10' , con)
```

####8. Which type of complaint was most common? 

```python
#YOUR CODE HERE
pd.read_sql( 'SELECT COUNT( request_type), request_type FROM sf_311'
           ' GROUP BY request_type '
           ' ORDER BY COUNT( request_type) DESC' 
           , con)
```

####9. Are there any neighborhoods where sidewalk cleaning requests aren't the most common? 

```python
#YOUR CODE HERE
pd.read_sql( 'SELECT neighborhood, request_type, COUNT(request_type) FROM sf_311'
            ' GROUP BY neighborhood, request_type '
            'ORDER BY count(*) DESC ' , con)
```
