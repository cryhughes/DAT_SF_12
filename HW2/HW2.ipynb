```python
#import (from 06-Neighbor Lab)
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets, feature_selection
from sklearn.cross_validation import train_test_split, cross_val_score
```

```python
#create params for train/test
feature_cols = [col for col in df.columns if col not in 'Wine']
x = df.ix[:, feature_cols]
feature_cols

#df.ix?
```

```python

# Create the training (and test) set using scikit-learn's train_test_split function
X_train, X_test, y_train, y_test = train_test_split( x , df['Wine'], 
                                                    test_size=0.3, random_state=42)

#X_train, X_test, y_train, y_test = train_test_split?
'''
X_train, X_test, y_train, y_test = train_test_split( iris.data , iris.target, test_size=0.3, 
                                                    random_state=42)
'''
```

```python
#define neighbors
# Loop through each neighbors value from 1 to 51 and append
# the scores
n_neighbors = range(1, 51, 2)
scores = []

for n in n_neighbors:
    clf = neighbors.KNeighborsClassifier(n)
    clf.fit(X_train, y_train)
    scores.append(clf.score(X_test, y_test))

#plot
plt.figure(figsize=(18,6))
_ = plt.plot(n_neighbors, scores, linewidth=3.0)
```

```python
from sklearn.cross_validation import cross_val_score

#run model with n = max, eg 27
clf = neighbors.KNeighborsClassifier( 27, weights='uniform')
scores = cross_val_score( clf, x, df['Wine'], cv=5)
print scores

#neighbors.KNeighborsClassifier? #weights = 'distance' looks appealing

#cross_val_score?
'''
cross_val_score(estimator, X, y=None, scoring=None, cv=None, n_jobs=1, verbose=0, fit_params=None, score_func=None, pre_dispatch='2*n_jobs') 
estimator <- model
x: data to fit
y: target
cv: number of x-validation folds
'''
```

```python
# perform clustering and plot the dendrogram
from scipy.cluster.hierarchy import linkage, dendrogram

#R = dendrogram(linkage(distx, method='single'), color_threshold=10)
R = dendrogram(linkage(distx, method='single')) # color_threshold)


plt.xlabel('points')
plt.ylabel('Height')
plt.suptitle('Cluster Dendrogram', fontweight='bold', fontsize=14);
```
